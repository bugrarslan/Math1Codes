```python
import sympy as sp # import sympy library
x = sp.Symbol('x')

def r(x):
    return x**2 + 1
    
def p(x):
    return 3*x + 2
    
sp.expand(r(p(x))) # (rop)(x)
```

```python
p(r(x)) # (por)(x)
```
