```python
def control(x, y):
  if(x > y):
    print(x, "sayısı", y, "sayısından büyüktür")
  elif(x < y):
    print(x, "sayısı", y, "sayısından küçüktür")
  else:
    print(x, "sayısı", y, "sayısına eşittir")
```

```python
control(9,2)
```

```python
control(2,9)
```

```python
control(9,9)
```

```python
def ciftVeOndanBuyuk(a):
  if(a % 2 == 0 and a > 10):
    print(a, "sayısı çift ve 10\'dan büyüktür")
  elif(a % 2 == 0 and a < 10):
    print(a, "sayısı çift fakat 10\'dan küçüktür")
  elif(a % 2 == 0 and a == 10):
    print(a, "sayısı çift fakat 10\'dan büyüktür")
  else:
    print(a, "sayısı çift ve 10 dan büyük değildir")
```

```python
ciftVeOndanBuyuk(12)
```

```python
ciftVeOndanBuyuk(10)
```

```python
ciftVeOndanBuyuk(13)
```

```python
ciftVeOndanBuyuk(9)
```
